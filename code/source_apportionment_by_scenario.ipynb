```python
import pandas as pd
```

```python
scens = ["Baseline", "Scenario_A", "Scenario_B"]
areas = ["Indre Oslofjord", "Ytre Oslofjord", "Hele Oslofjord"]
pars = ["TOTN", "DIN", "TON", "TOTP", "TDP", "TPP", "TOC", "SS"]

csv_path = r"../data/results_summary.csv"
df = pd.read_csv(csv_path)

df_list = []
for scen in scens:
    for area in areas:
        for par in pars:
            par_df = df.query(
                "(`Område` == @area) and (Scenario == @scen) and (Parameter == @par)"
            ).copy()
            total = par_df["Verdi (tonn)"].sum()
            par_df["Prosent"] = (par_df["Verdi (tonn)"] / total) * 100

            cols = ["Område", "Scenario", "Kilde", "Parameter", "Prosent"]
            par_df = par_df[cols]

            df_list.append(par_df)
df = pd.concat(df_list, axis="rows").reset_index(drop=True)

df["Kilde"] = df["Kilde"] + " (%)"
df["Prosent"] = df["Prosent"].round(2)

df = df.pivot(
    index=["Område", "Scenario", "Parameter"], columns="Kilde", values="Prosent"
).reset_index()
df = df.fillna(0)

df.to_excel(r"../data/source_apportionment_by_scenario.xlsx", index=False)

df.head()
```
