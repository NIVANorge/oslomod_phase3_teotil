```python
import pandas as pd
```

```python
header = [
    "regine",
    "TOTN_agri",
    "TOTN_back",
    "DIN_agri",
    "DIN_back",
    "TOTP_agri",
    "TOTP_back",
    "TDP_agri",
    "TDP_back",
    "SS_agri",
    "SS_back",
    "TOC_agri",
    "TOC_back",
]
pars = ["TOTN", "DIN", "TOTP", "TDP", "SS", "TOC"]
df_list = []
scenarios = ["Baseline", "SCA", "SCB"]
for year in range(2017, 2020):
    for scen in scenarios:
        xl_path = f"/home/jovyan/shared/common/oslofjord_modelling/phase3_scenarios/agri_raw/agri_data_v20250912/Leveranse_{year}.xlsx"
        df = pd.read_excel(
            xl_path, sheet_name=scen, header=None, names=header, skiprows=2
        )
        df["scenario"] = scen
        df["year"] = year
        df_list.append(df)
df = pd.concat(df_list, axis="rows")
for par in pars:
    df[par] = df[f"{par}_agri"] #+ df[f"{par}_back"]
    del df[f"{par}_agri"], df[f"{par}_back"]
df.head()
```

```python
df = df.groupby("scenario").sum(numeric_only=True)
df
```

```python
df / 3
```

```python
for par in pars:
    base = df.loc["Baseline", par]
    sca = df.loc["SCA", par]
    scb = df.loc["SCB", par]

    sca_pct = 100 * (sca - base) / base
    scb_pct = 100 * (scb - base) / base

    print("")
    print(f"# {par} #######################")
    print(f"SCA: {(sca - base)/1000:.3f} tonnes")
    print(f"SCA: {sca_pct:.1f} %")
    print(f"SCB: {(scb - base)/1000:.3f} tonnes")
    print(f"SCB: {scb_pct:.1f} %")
```
